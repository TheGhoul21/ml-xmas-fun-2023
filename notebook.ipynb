```python
import numpy as np

def softmax(x):
    return(np.exp(x)/np.exp(x).sum())

def softmax_derivative(x):
    return softmax(x) * (1 - softmax(x))


z = np.linspace(-5, 5, 100)

import matplotlib.pyplot as plt
softmax_values = np.array([softmax([zi, 0])[0] for zi in z])  # Taking softmax of two-element vector for visualization
softmax_derivative_values = np.array([softmax_derivative([zi, 0])[0] for zi in z])



# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(z, softmax_values, label='Softmax')
plt.title('Softmax Function')
plt.xlabel('z')
plt.ylabel('softmax(z)')

plt.subplot(1, 2, 2)
plt.plot(z, softmax_derivative_values, label='Derivative', color='orange')
plt.title('Derivative of Softmax')
plt.xlabel('z')
plt.ylabel('d(softmax)/dz')

plt.tight_layout()
plt.show()


```
